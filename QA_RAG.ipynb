```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
# Load the document (a long text file)
with open('/content/The Grand Design.pdf', 'rb') as file: # Open the file in binary mode ('rb')
    document = file.read()

# Decode the binary content using a suitable encoding (e.g., latin-1)
document = document.decode('latin-1')

# Split the document into chunks using a text splitter (adjust chunk size based on the dataset)
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100,)

chunks = text_splitter.split_text(document)

print(f"Total number of chunks: {len(chunks)}")
```

```python
from sentence_transformers import SentenceTransformer
import torch
import time
start_time = time.time()
# Use a faster embedding model
# Load the model on the GPU
embedding_model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens', device='cuda' if torch.cuda.is_available() else 'cpu')


# Batch processing for generating embeddings
batch_size = 128  # Adjust based on your memory limits
embeddings = []
for i in range(0, len(chunks), batch_size):
    batch_embeddings = embedding_model.encode(chunks[i:i+batch_size], show_progress_bar=True)
    embeddings.append(batch_embeddings)



```

```python
import numpy as np
# Convert list to numpy array
embeddings = np.vstack(embeddings)



print(f"Generated embeddings for {len(embeddings)} chunks.")
```

```python
import faiss

# Get the dimension of the embeddings
dimension = embeddings.shape[1]

# Initialize the FAISS index
index = faiss.IndexFlatL2(dimension)

# Add the embeddings to the FAISS index
index.add(embeddings)

print(f"Indexed {index.ntotal} document chunks in FAISS.")
```

```python
def retrieve_relevant_chunks(query, k=5):
    # Generate the embedding for the query using the same embedding model
    query_embedding = embedding_model.encode([query])

    # Search in the FAISS index for top-k relevant chunks
    distances, indices = index.search(np.array(query_embedding), k)

    # Retrieve the corresponding chunks based on the indices
    relevant_chunks = [chunks[i] for i in indices[0]]

    return relevant_chunks
```

```python
query = "What is M-theory according to Stephen Hawking?"
relevant_chunks = retrieve_relevant_chunks(query, k=5)

# Print the relevant chunks to see what has been retrieved
print("Relevant Chunks:")
for chunk in relevant_chunks:
    print(chunk)
```

```python
from transformers import pipeline

# Load the Hugging Face generative model
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-1.3B',device=0)

# Combine the relevant chunks to create the context
context = " ".join(relevant_chunks)

# Create the prompt for the generative model
prompt = f"Context: {context}\nQuestion: {query}\nAnswer:"

# Generate the answer using the model
response = generator(prompt, max_new_tokens=150, truncation=True)

# Output the generated answer
print("Generated Answer:", response[0]['generated_text'])
```
